```python
"""
Limpiamos la columna de genre haciendo replace en alguna fila que ponia Dark Wave' por 'New Wave'
"""
```

```python
""" 
Quitamos los corchetes de la columna templo
"""
```

```python
df_musica['tempo'] = df_musica['tempo'].str.strip('[]').astype(float)
df_musica['tempo']
```

```python
df_musica
```

```python
columnas_a_formatear = ['rmse_1', 'rmse_2', 'rmse_3', 'rmse_4', 'rmse_5','rmse_6', 'rmse_7', 'rmse_8', 'rmse_9', 'rmse_10', 'rmse_11', 'rmse_12', 'rmse_13', 'rmse_14', 'rmse_15', 
                        'rmse_16', 'rmse_17', 'rmse_18', 'rmse_19', 'rmse_20']
# Verifica si las columnas existen en el DataFrame
columnas_presentes = [col for col in columnas_a_formatear if col in df_musica.columns]

# Aplicar formato con dos decimales
df_musica[columnas_presentes] = df_musica[columnas_presentes].applymap(lambda x: f"{x:.8f}")

# Guardar el archivo con los valores formateados
df_musica.to_csv('musicafinal.csv', index=False)
```
