```python
import librosa

def extract_features(y, sr):
    # Calcular el tempo y los beats
    tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
    beats_per_minute = len(beat_frames) * (60 / (len(y) / sr))  # Beats por minuto

    # Calcular características espectrales y de frecuencia
    rmse = librosa.feature.rms(y=y).mean()  # Promedio del ruido (RMSE)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)  # 20 coeficientes MFCC
    mfcc_means = mfcc.mean(axis=1)  # Promedios por cada MFCC

    zcr = librosa.feature.zero_crossing_rate(y).mean()
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr).mean()
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr).mean()
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr).mean()
    chroma = librosa.feature.chroma_stft(y=y, sr=sr).mean()

    # Crear un diccionario con los valores extraídos
    features = {
        "beats": beats_per_minute,
        "tempo": tempo,
        "rmse": rmse,
        "mfcc_1": mfcc_means[0] if len(mfcc_means) > 0 else None,
        "mfcc_2": mfcc_means[1] if len(mfcc_means) > 1 else None,
        "mfcc_3": mfcc_means[2] if len(mfcc_means) > 2 else None,
        "mfcc_4": mfcc_means[3] if len(mfcc_means) > 3 else None,
        "mfcc_5": mfcc_means[4] if len(mfcc_means) > 4 else None,
        "mfcc_6": mfcc_means[5] if len(mfcc_means) > 5 else None,
        "mfcc_7": mfcc_means[6] if len(mfcc_means) > 6 else None,
        "mfcc_8": mfcc_means[7] if len(mfcc_means) > 7 else None,
        "mfcc_9": mfcc_means[8] if len(mfcc_means) > 8 else None,
        "mfcc_10": mfcc_means[9] if len(mfcc_means) > 9 else None,
        "mfcc_11": mfcc_means[10] if len(mfcc_means) > 10 else None,
        "mfcc_12": mfcc_means[11] if len(mfcc_means) > 11 else None,
        "mfcc_13": mfcc_means[12] if len(mfcc_means) > 12 else None,
        "mfcc_14": mfcc_means[13] if len(mfcc_means) > 13 else None,
        "mfcc_15": mfcc_means[14] if len(mfcc_means) > 14 else None,
        "mfcc_16": mfcc_means[15] if len(mfcc_means) > 15 else None,
        "mfcc_17": mfcc_means[16] if len(mfcc_means) > 16 else None,
        "mfcc_18": mfcc_means[17] if len(mfcc_means) > 17 else None,
        "mfcc_19": mfcc_means[18] if len(mfcc_means) > 18 else None,
        "mfcc_20": mfcc_means[19] if len(mfcc_means) > 19 else None,
        "zero_crossing_rate": zcr,
        "rolloff": rolloff,
        "spectral_centroid": spectral_centroid,
        "spectral_bandwidth": spectral_bandwidth,
        "chroma_stft": chroma
    }

    return features

# Lista para almacenar los resultados
datos_audio = []

# Procesar cada archivo en la lista
for ruta in rutas_audio:
    try:
        y, sr = librosa.load(ruta, sr=None)  # Cargar archivo de audio
        caracteristicas = extract_features(y, sr)  # Extraer características
        caracteristicas["file"] = ruta  # Agregar el nombre del archivo
        datos_audio.append(caracteristicas)  # Guardar los resultados
    except Exception as e:
        print(f"Error procesando {ruta}: {e}")

# Crear un DataFrame con los datos
df_audio = pd.DataFrame(datos_audio)

# Mostrar el DataFrame
print(df_audio)

# Opcional: Guardar el DataFrame en un archivo CSV
df_audio.to_csv("features_audio.csv", index=False)
```
