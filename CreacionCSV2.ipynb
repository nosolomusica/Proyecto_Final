```python
def extract_features(y, sr):
    tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
    beats = len(beat_frames)
    rmse = librosa.feature.rms(y=y)
    rmse_values = rmse[0][:20]  # Tomamos los primeros 20 valores de RMSE
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13).mean()
    zcr = librosa.feature.zero_crossing_rate(y).mean()
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr).mean()
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr).mean()
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr).mean()
    chroma = librosa.feature.chroma_stft(y=y, sr=sr).mean()


    return {
        "beats": beats,
        "tempo": tempo,
        "mfcc": mfcc.tolist(),
        "zero_crossing_rate": zcr,
        "rolloff": rolloff,
        "spectral_centroid": spectral_centroid,
        "spectral_bandwidth": spectral_bandwidth,
        "chroma_stft": chroma,
        "rmse_1": rmse_values[0] if len(rmse_values) > 0 else None,
        "rmse_2": rmse_values[1] if len(rmse_values) > 1 else None,
        "rmse_3": rmse_values[2] if len(rmse_values) > 2 else None,
        "rmse_4": rmse_values[3] if len(rmse_values) > 3 else None,
        "rmse_5": rmse_values[4] if len(rmse_values) > 4 else None,
        "rmse_6": rmse_values[5] if len(rmse_values) > 5 else None,
        "rmse_7": rmse_values[6] if len(rmse_values) > 6 else None,
        "rmse_8": rmse_values[7] if len(rmse_values) > 7 else None,
        "rmse_9": rmse_values[8] if len(rmse_values) > 8 else None,
        "rmse_10": rmse_values[9] if len(rmse_values) > 9 else None,
        "rmse_11": rmse_values[10] if len(rmse_values) > 10 else None,
        "rmse_12": rmse_values[11] if len(rmse_values) > 11 else None,
        "rmse_13": rmse_values[12] if len(rmse_values) > 12 else None,
        "rmse_14": rmse_values[13] if len(rmse_values) > 13 else None,
        "rmse_15": rmse_values[14] if len(rmse_values) > 14 else None,
        "rmse_16": rmse_values[15] if len(rmse_values) > 15 else None,
        "rmse_17": rmse_values[16] if len(rmse_values) > 16 else None,
        "rmse_18": rmse_values[17] if len(rmse_values) > 17 else None,
        "rmse_19": rmse_values[18] if len(rmse_values) > 18 else None,
        "rmse_20": rmse_values[19] if len(rmse_values) > 19 else None,    
        
    }

# Lista para almacenar los resultados
datos_audio3 = []

# Procesar cada archivo en la lista
for ruta in rutas_audio3:
    try:
        y, sr = librosa.load(ruta, sr=None)  # Cargar archivo de audio
        caracteristicas = extract_features(y, sr)  # Extraer características
        caracteristicas["file"] = ruta  # Agregar el nombre del archivo
        datos_audio3.append(caracteristicas)  # Guardar los resultados
    except Exception as e:
        print(f"Error procesando {ruta}: {e}")

# Crear un DataFrame con los datos
df_audio3 = pd.DataFrame(datos_audio3)

# Mostrar el DataFrame
print(df_audio3)

# Opcional: Guardar el DataFrame en un archivo CSV
df_audio3.to_csv("caracteristicas_audio3.csv", index=False)
```

```python
datos_canciones3 = []

# Procesar cada archivo en la lista
for ruta in rutas_audio3:
    # Obtener solo el nombre del archivo
    archivo = ruta.split('/')[-1]
    # Separar el nombre del archivo en partes (artista, canción, género, subgénero y año)
    partes = archivo.split('_')
    if len(partes) >= 5:  # Asegurarse de que hay al menos tres partes
        artist = partes[0].replace('-', ' ')  # Reemplazar guiones por espacios
        song = partes[1].replace('-', ' ')
        genre = partes[2].replace('-', ' ')  # Quitar la extensión
        subgenre = partes[3].replace('-', ' ')
        year = partes[4].replace('-', ' ').replace('.mp3', '')
        # Añadir los datos como diccionario a la lista
        datos_canciones3.append({
            "artist": artist,
            "song": song,
            "genre": genre,
            "subgenre": subgenre,
            "year": year
        })

# Crear un DataFrame con los datos
df_canciones3 = pd.DataFrame(datos_canciones3)

# Mostrar el DataFrame
print(df_canciones3)

# Opcional: Guardar el DataFrame en un archivo CSV
df_canciones3.to_csv("datos_canciones3.csv", index=False)
```
